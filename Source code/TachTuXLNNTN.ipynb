### Hàm xử lí dữ liệu đầu vào, tách câu thành các từ riêng biệt

```python
# Phân tách câu bằng khoảng trắng và các kí tự đặc biệt
def split_sentences(sentences):
    final_array = []
    cur = 0
    i = 0

    while i < len(sentences):
        # nếu phát hiện ra dấu câu thì tách
        if sentences[i] in special_character:
            # xử lí trường hợp tường thuật
            if sentences[i] == ":":
                if sentences[i + 1] == " ":
                    sentences = sentences[:i + 1] + sentences[i + 2:]
            # check xem dấu chấm than có nằm cuối câu không
            if not (sentences[i] == "!" and i != len(sentences) - 1):
                left = sentences[cur:i].strip().split(" ")

                if left[0] != "":

                    for item in left:
                        final_array.append(item)
                final_array.append(sentences[i])
                cur = i + 1
        i = i + 1
    # nếu không có dấu câu từ vị trí split trước đó về cuối
    if cur < len(sentences):
        left = sentences[cur:i].strip().split(" ")
        for item in left:
            final_array.append(item)
    return final_array
```

### Tách từ bằng phương pháp Maximum Matching

```python
# Nối từ bằng phương pháp Maximum Matching
def split_string(string):
    array = split_sentences(string)
    i = 0
    pos = 4
    final_arr = []
    while i < len(array):
        while True:
            cur_string = " ".join(array[i:i + pos])
            # khi chỉ còn 1 tiếng
            if pos == 1:
                break
            # Nếu chứa kí tự đặc biệt thì bỏ qua
            if not re.search("[,()–:;.?!\"]", cur_string):
                # Check Danh từ riêng, tất cả các tiếng bắt đầu bằng chữ cái viết hoa và không chứa số
                if cur_string.istitle() and not re.search("\d", cur_string):
                    # check xem có nằm đầu dòng không. Hoặc nằm sau dấu  ? or !
                    if i == 0:
                        # nếu tiếng nằm đầu dòng có trong từ điển thì tách ra. tuy nhiên cách này cũng k bao phủ tất cả trường hợp, ví dụ trong tên riêng từ đầu có trong từ điển
                        if array[i].lower() in dictionary:
                            final_arr.append(array[i])
                            cur_string = " ".join(array[i + 1:i + pos])

                    i = i + pos - 1
                    break
                # check xem tiếng có nằm trong từ điển
                if cur_string.lower() in dictionary:
                    i = i + pos - 1
                    break
            pos = pos - 1
        final_arr.append(cur_string.replace(" ", "_"))
        pos = 4
        i = i + 1
    return final_arr
```

# Tách từ Bằng WFST

### Hàm Tách từ trong bộ train

```python
def split_sentences(_sentences):
    _sentences = _sentences.strip().split()
    if "" in _sentences:
        _sentences.remove("")
    return _sentences
```

### Hàm tách từ trong một câu

```python
def split_special_character(sentences):
    final_array = []
    cur = 0
    i = 0
    while i < len(sentences):
        # nếu phát hiện ra dấu câu thì tách
        if sentences[i] in special_character:
            # xử lí trường hợp tường thuật
            if sentences[i] == ":":
                if sentences[i + 1] == " ":
                    sentences = sentences[:i + 1] + sentences[i + 2:]
            # check xem dấu chấm than có nằm cuối câu không
            if not (sentences[i] == "!" and i != len(sentences) - 1):
                left = sentences[cur:i].strip().split(" ")
                # trường hợp dấu câu dạng :"
                if left[0] != "":
                    final_array.append(left)
                final_array.append(sentences[i])
                cur = i + 1
        i = i + 1
    # nếu không có dấu câu ở cuối
    if cur < len(sentences):
        left = sentences[cur:i].strip().split(" ")
        final_array.append(left)
    return final_array
```

### Tách từ bằng WFST

```python
# tìm tất cả các trường hợp có thể tách của một câu
def all_possible_sentences(test_case, i, array, length_case, S):
    if i + 4 > len(test_case):
        end = len(test_case)
    else:
        end = i + 4
    for j in range(i, end):
        cur_words = "_".join(test_case[i:j + 1])

        if cur_words in dictionary or j - i == 0:
            # check trường hợp viết hoa, tương tự như phương pháp mxm matching check xem từ đầu có nằm trong từ điển và không chứa số
            if not (cur_words.istitle() and test_case[i].lower() in dictionary and re.search("\d",
                                                                                             cur_words)):
                S.append(cur_words)

                length_case += j - i + 1
                if length_case < len(test_case):
                    all_possible_sentences(test_case, j + 1, array, length_case, S)
                else:
                    array.append(S.copy())

                S.pop()
                length_case -= j - i + 1
    return array
```

```python
test_case="riêng tư cách của anh đã không đủ rồi !"
test_case= test_case.split(" ")
all_possible_sentences(test_case, 0, [], 0, [])
```

```python
# Tìm tổng xác suất của câu có giá trị nhỏ nhất
def split_single_sentences(test_case):
    final_arr = all_possible_sentences(test_case, 0, [], 0, [])

    min_value = sys.maxsize
    final_sentences = []
    for item in final_arr:
        value = 0
        for word in item:
          #lấy xác suất từ trong từ điển ra
            if word in dictionary:
                value += dictionary[word]
            else:    
                value += -Decimal(math.log2(1 / len_dictionary))
        if value < min_value:
            min_value = value
            final_sentences = item
    
    return final_sentences
```

```python
test_case="riêng tư cách của anh đã không đủ rồi !"
test_case= test_case.split(" ")
split_single_sentences(test_case)
```

```python
#Hàm xử lí giá trị đầu vào của câu cũng như dùng tất cả các hàm trên để tạo câu hoàn chỉnh
def split_sen(_sen):
    len_dictionary = len(dictionary)
    input_test_case = _sen
    # đầu tiên tách câu thành các từ riêng biệt
    input_test_case = split_special_character(input_test_case)

    # Xử lí in hoa ở đầu câu
    if len(input_test_case[0]) > 1:
        if not input_test_case[0][1].istitle():
            input_test_case[0][0] = input_test_case[0][0][0].lower() + input_test_case[0][0][1:]
    #Xử lí in hoa sau :" 
    for i in range(1, len(input_test_case) - 1):
        if input_test_case[i] == '''"''' and input_test_case[i - 1] == ":":
            if not input_test_case[i + 1][1].istitle():
                input_test_case[i + 1][0] = input_test_case[i + 1][0][0].lower() + input_test_case[i + 1][0][1:]
    
    #xử lí input_test_case cho ra câu có xác suất cao nhất
    result = []
    for item in input_test_case:
        if isinstance(item, list):
            cur_arr = split_single_sentences(item)
            for i in cur_arr:
                result.append(i)
        else:
            result.append(item)

    # Trả lại in hoa chữ cái đầu
    result[0] = result[0][0].upper() + result[0][1:]
    for i in range(1, len(result) - 1):
        if result[i] == '''"''' and result[i - 1] == ":":
            result[i + 1] = result[i + 1][0].upper() + result[i + 1][1:]
    return result
```

###So sánh các phương pháp

```python
#Phân tách câu thành các từ riêng biệt
def split_compare_sentences(_sen):
    array = []
    cur_pos = 0
    i = 0
    while i < len(_sen):
        if _sen[i] == " ":
            array.append(_sen[cur_pos: i])
            cur_pos = i
        elif _sen[i] == "_":
            array.append(_sen[cur_pos: i])
            cur_pos = i
        i += 1
    array.append(_sen[cur_pos:])
    return array
```

```python
#ở đây e chia câu thành các từ riêng biệt để so sánh dựa trên index và phân cách trước nó: nếu có _ thì nó là bộ phận của từ ghép 
split_compare_sentences("Tôi khuyên_giáo sư thay_đổi cách nghĩ của mình về vấn_đề này")
```

```python
#Hàm so sánh 2 câu 
def result_compare_sentences(sentences1, sentences2):
    sentences1 = split_compare_sentences(sentences1)
    sentences2 = split_compare_sentences(sentences2)
    i = 0
    result = 0
    while i < len(sentences1):
      #kiểm tra xem có dấu _ trước không
        if sentences1[i][0] == "_":
          #chắc chắc trước nó là một từ trong từ ghép
            cur_sentences = [sentences1[i - 1], sentences1[i]]
            if i + 1 < len(sentences1):
                #check xem sau nó có phải là một từ trong từ ghép của nó không
                if sentences1[i + 1][0] == "_":
                    cur_sentences.append(sentences1[i + 1])
                    if i + 2 < len(sentences1):
                        if sentences1[i + 2][0] == "_":
                            cur_sentences.append(sentences1[i + 2])

            compare_sentences = []
            #so sánh dựa trên index
            for j in range(i - 1, i - 1 + len(cur_sentences)):
                compare_sentences.append(sentences2[j])

            if cur_sentences == compare_sentences:
                result += 1
            i += len(cur_sentences) - 1
        i += 1
    return result
```

```python
sen1="Tôi khuyên_giáo sư thay_đổi cách nghĩ của mình về vấn_đề này"
sen2="Tôi khuyên giáo_sư thay_đổi cách nghĩ của mình về vấn_đề này"
result_compare_sentences(sen1, sen2)
```

###Chạy so sánh trên toàn tập dữ liệu

```python
sum_score_vncore = 0
sum_score_tachtu_MXM = 0
sum_score_tachtu_WFST = 0
for i in range(0, len(GoldResult)):
    sum_score_vncore += result_compare_sentences(GoldResult[i], VnCoreNlpResult[i])
    sum_score_tachtu_MXM += result_compare_sentences(GoldResult[i], TachTuMXM[i])
    sum_score_tachtu_WFST += result_compare_sentences(GoldResult[i], TachTuWFST[i])

max_score = 0
for item in GoldResult:
    item = item.split(" ")
    for j in item:
        if j.count("_") > 0:
            max_score += 1 
count_word= 0
for item in GoldResult:
  item = item.split(" ")
  for sentence in item:
    A = len(sentence)
    count_word = count_word + A

print('- Maximum Matching: '+"Accuracy = " + str(sum_score_tachtu_MXM / max_score * 100))
print('- WFST: '+ "Accuracy = " + str((sum_score_tachtu_WFST / max_score) * 100))
print('- VncoreNLP: '+"Accuracy = " + str(sum_score_vncore / max_score * 100))
```
